```python
def download(url,dir): 
  name = random.randrange(1, 99999) 
  file_name = dir+"/"+ str(name) + ".mp3"     
  with open(file_name, "wb") as file:
    response = get(url)
    file.write(response.content)
```

```python
for x,y in zip(train_df['sp'],train_df['file']):
    if x == 'sonneratii':
        path = 'SouthAsia_Recordings/sonneratii'
        download(y,path)
    elif x=='pondicerianus':
        path = 'SouthAsia_Recordings/pondicerianus'
        download(y,path)
    elif x=='torqueola':
        path = 'SouthAsia_Recordings/torqueola'
        download(y,path)
    elif x=='cristatus':
        path = 'SouthAsia_Recordings/cristatus'
        download(y,path)
    elif x=='spadicea':
        path = 'SouthAsia_Recordings/spadicea'
        download(y,path)
    elif x=='impejanus':
        path = 'SouthAsia_Recordings/impejanus'
        download(y,path)
    elif x=='gallus':
        path = 'SouthAsia_Recordings/gallus'
        download(y,path)
    elif x=='francolinus':
        path = 'SouthAsia_Recordings/francolinus'
        download(y,path)
    elif x=='lafayettii':
        path = 'SouthAsia_Recordings/lafayettii'
        download(y,path)
    elif x=='bicalcarata':
        path = 'SouthAsia_Recordings/bicalcarata'
        download(y,path)
    elif x=='asiaticus':
        path = 'SouthAsia_Recordings/asiaticus'
        download(y,path)
    elif x=='ceylonensis':
        path = 'SouthAsia_Recordings/ceylonensis'
        download(y,path)
    elif x=='melanurus':
        path = 'SouthAsia_Recordings/melanurus'
        download(y,path)
    else:
        print("Invalid Species")
```
