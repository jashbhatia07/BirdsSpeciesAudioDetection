```python
import urllib.request
import wave

# URL of the music file to download
url = 'https://xeno-canto.org/569020/download'

# Open the URL and download the music file
response = urllib.request.urlopen(url)

# Open a wave file to write the music to
with wave.open('music1.wav', 'wb') as f:
    # Set the wave file parameters
    f.setnchannels(2)
    f.setsampwidth(2)
    f.setframerate(44100)

    # Write the music data to the wave file
    f.writeframes(response.read())

# Close the response and the wave file
response.close()
f.close()
```

```python
from requests import get                        #to make GET request 
import random 
def download(url): 
  name = random.randrange(1, 99999) 
  file_name = str(name) + ".mp3"     
  with open(file_name, "wb") as file:  	# open in binary mode 
    response = get(url)                     # get request 
    file.write(response.content)            # write to file 
download("https://xeno-canto.org/569020/download")
```
