# India

```python
import pandas as pd
import requests
```

```python
d = dict()
```

```python
def fetch_indianData_AllSpecies(d):
    for i in range(1,34):
        response = requests.get(f'https://xeno-canto.org/api/2/recordings?query=cnt:india&page={i}')
        df = pd.DataFrame(response.json()['recordings'])
        d[f'df{i}'] = df
    return d
```

```python
d = fetch_indianData_AllSpecies(d)
```

```python
df = pd.concat([d['df1'], d['df2'],d['df3'],d['df4'],d['df5'],d['df6'],d['df7'],d['df8'],d['df9'],d['df10'],d['df11'], d['df12'], d['df13'], d['df14'], d['df15'], d['df16'], d['df17'], d['df18'], d['df19'], d['df20'], d['df21'], d['df22'], d['df23'], d['df24'], d['df25'], d['df26'], d['df27'], d['df28'], d['df29'], d['df30'], d['df31'], d['df32'], d['df33']], axis=0)
```
