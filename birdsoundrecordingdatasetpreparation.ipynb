```python
def download(url,dir): 
  name = random.randrange(1, 99999) 
  file_name = dir+"/"+ str(name) + ".mp3"     
  with open(file_name, "wb") as file:
    response = get(url)
    file.write(response.content)
```

```python
for x,y in zip(train_df['sp'],train_df['file']):
    if x == 'sonneratii':
        path = 'Indian_Recordings/sonneratii'
        download(y,path)
    elif x=='pondicerianus':
        path = 'Indian_Recordings/pondicerianus'
        download(y,path)
    elif x=='torqueola':
        path = 'Indian_Recordings/torqueola'
        download(y,path)
  
    elif x=='spadicea':
        path = 'Indian_Recordings/spadicea'
        download(y,path)
    elif x=='impejanus':
        path = 'Indian_Recordings/impejanus'
        download(y,path)
    elif x=='gallus':
        path = 'Indian_Recordings/gallus'
        download(y,path)
    elif x=='francolinus':
        path = 'Indian_Recordings/francolinus'
        download(y,path)
    elif x=='lafayettii':
        path = 'Indian_Recordings/pictus'
        download(y,path)
    elif x=='bicalcarata':
        path = 'Indian_Recordings/coromandelica'
        download(y,path)
    elif x=='asiaticus':
        path = 'Indian_Recordings/macrolopha'
        download(y,path)
    else:
        print("Invalid Species")
```
