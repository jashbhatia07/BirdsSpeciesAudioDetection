```python
import os
from requests import get
import random 
```

```python
os.mkdir('/kaggle/working/sonneratii')
os.mkdir('/kaggle/working/pondicerianus')
os.mkdir('/kaggle/working/torqueola')
os.mkdir('/kaggle/working/cristatus')
os.mkdir('/kaggle/working/spadicea')
os.mkdir('/kaggle/working/impejanus')
```

```python
def download(url,dir): 
  name = random.randrange(1, 99999) 
  file_name = dir+"/"+ str(name) + ".mp3"     
  with open(file_name, "wb") as file:
    response = get(url)
    file.write(response.content)
```

```python
for x,y in zip(train_df['sp'],train_df['file']):
    if x == 'sonneratii':
        path = '/kaggle/working/sonneratii'
        download(y,path)
    elif x=='pondicerianus':
        path = '/kaggle/working/pondicerianus'
        download(y,path)
    elif x=='torqueola':
        path = '/kaggle/working/torqueola'
        download(y,path)
    elif x=='cristatus':
        path = '/kaggle/working/cristatus'
        download(y,path)
    elif x=='spadicea':
        path = '/kaggle/working/spadicea'
        download(y,path)
    elif x=='impejanus':
        path = '/kaggle/working/impejanus'
        download(y,path)
    else:
        print("Invalid Species")
        
```
